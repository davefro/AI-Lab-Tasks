```python
# Import the load_digits to load the handwritten digits dataset
from sklearn.datasets import load_digits
```

```python
# Load the digits dataset
digits = load_digits()
```

```python
# Import the train_test_split function 
from sklearn.model_selection import train_test_split
# Splits the datasets
X_train, X_test, y_train, y_test = train_test_split(
    digits.data, digits.target, random_state=11, test_size=0.20)
```

```python
# import KNeighborsClassifier 
from sklearn.neighbors import KNeighborsClassifier
# import svc
from sklearn.svm import SVC
#  import GaussianNB
from sklearn.naive_bayes import GaussianNB
```

```python
# define a dictionary 'estimators' containing three classifiers
estimators = {
        'KNeighborsClassifier': KNeighborsClassifier(),
        'SVC': SVC(gamma='scale'),
        'GaussianNB': GaussianNB()}
```

```python
# import KFold to split data in train/test sets
from sklearn.model_selection import KFold
```

```python
# create a KFold object for 10-ford cross validation
kfold = KFold(n_splits=10, random_state=11, shuffle=True)
```

```python
# import cross_val_score to evaluate a model's performance
from sklearn.model_selection import cross_val_score
```

```python
# iterate over each classifier and evaluate each one
for estimator_name, estimator_object in estimators.items():
      kfold = KFold(n_splits=10, random_state=11, shuffle=True)

      scores = cross_val_score(estimator=estimator_object,X=digits.data, y=digits.target, cv=kfold)

      print(f'{estimator_name:>20}: ' + f'mean accuracy={scores.mean():.2%}; ' + f'standard deviation={scores.std():.2%}')
```
